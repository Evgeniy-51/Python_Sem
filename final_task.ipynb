```python
import sympy as sp
```

```python
x = sp.Symbol("x")
```

```python
f = x**3 - 50*x
f
```

```python
g = -x**4 + 88*x**2 - 241
g
```

```python
eq = sp.Eq(x**3 - 50*x, -x**4 + 88*x**2 - 241)
res = sp.solveset(eq)
```

```python
print(f'Всего точек пересечения: {len(res)}')
res
```

```python
sp.plot(f,g)
```

```python
fg = f - g
fg
```

```python
sp.plot(fg)
```
